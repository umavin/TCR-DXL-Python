```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
l4 = np.array([100,200,300,300,300,300,300])
l5 = np.array([100,200,350,350,350,350,350])
l6 = np.array([100,200,350,360,360,360,360])

L_str = l4[0:2]

l4_cp = l4[2:4]
l5_cp = l5[2:4]
l6_cp = l6[2:4]

#print(l6_cp)


d = 30.000000000
k2 = np.divide(np.sqrt(l4_cp*l4_cp+l5_cp*l5_cp+l6_cp*l6_cp-l6_cp*l4_cp-l5_cp*l6_cp-l6_cp*l4_cp)*2.000000000,d*(l4_cp+l5_cp+l6_cp))
theta2 = np.divide(np.sqrt(l4_cp*l4_cp+l5_cp*l5_cp+l6_cp*l6_cp-l6_cp*l4_cp-l5_cp*l6_cp-l6_cp*l4_cp)*2.000000000,d*3.00000000000000)
phi2 = np.arctan2(np.sqrt(3)*(l5_cp+l6_cp-2*l4_cp),2*(l5_cp-l6_cp))

dL2 = np.divide(theta2,k2)

l1 = dL2 - theta2*d*np.cos(phi2+np.deg2rad(60.000))
l2 = dL2 - theta2*d*np.cos(phi2+np.deg2rad(300.000))
l3 = dL2 - theta2*d*np.cos(phi2+np.deg2rad(180.000))



l1l = l1[-1]
l2l = l2[-1]
l3l = l3[-1]

l1 = np.hstack((L_str,l1,[l1l+100,l1l+250,l1l+270]))
l2 = np.hstack((L_str,l2,[l2l+100,l2l+250,l2l+250]))
l3 = np.hstack((L_str,l3,[l3l+100,l3l+200,l3l+200]))
print(l1,l2,l3)
```

```python
plt.figure(1)
plt.plot(l1,label='L1')
plt.plot(l2,label='L2')
plt.plot(l3,label='L3')


plt.plot(l4,label='L4')
plt.plot(l5,label='L5')
plt.plot(l6,label='L6')

plt.legend()

plt.plot
```
